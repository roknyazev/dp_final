```python
import matplotlib.pyplot as plt
import pickle
import matplotlib

mpl_params = {

    'figure.figsize': (25, 13),
}
plt.rcParams.update(mpl_params)
font = {'weight' : 'bold',
        'size'   : 18}

matplotlib.rc('font', **font)
```

```python
f = open('results/genetic.pkl', 'rb')
results_genetic = pickle.load(f)
f.close()

prices_genetic = []
orders_count_genetic = []
prices_per_order_genetic = []
times_genetic = []
for result in results_genetic:
    prices_genetic.append(result.total_price)
    orders_count_genetic.append(len(result.orders))
    prices_per_order_genetic.append(result.total_price / len(result.orders))
    times_genetic.append(result.time)
```

```python
f = open('results/greedy1.pkl', 'rb')
results_greedy1 = pickle.load(f)
f.close()

prices_greedy1 = []
orders_count_greedy1 = []
prices_per_order_greedy1 = []
times_greedy1 = []
for result in results_greedy1:
    prices_greedy1.append(result.total_price)
    orders_count_greedy1.append(len(result.orders))
    prices_per_order_greedy1.append(result.total_price / len(result.orders))
    times_greedy1.append(result.time)
```

```python
f = open('results/greedy2.pkl', 'rb')
results_greedy2 = pickle.load(f)
f.close()

prices_greedy2 = []
orders_count_greedy2 = []
prices_per_order_greedy2 = []
times_greedy2 = []
for result in results_greedy2:
    prices_greedy2.append(result.total_price)
    orders_count_greedy2.append(len(result.orders))
    prices_per_order_greedy2.append(result.total_price / len(result.orders))
    times_greedy2.append(result.time)
```

```python
f = open('results/trivial.pkl', 'rb')
results_trivial = pickle.load(f)
f.close()

prices_trivial = []
orders_count_trivial = []
prices_per_order_trivial = []
times_trivial = []
for result in results_trivial:
    prices_trivial.append(result.total_price)
    orders_count_trivial.append(len(result.orders))
    prices_per_order_trivial.append(result.total_price / len(result.orders))
    times_trivial.append(result.time)
```

```python
plt.plot(orders_count_trivial, prices_per_order_trivial)
plt.scatter(orders_count_trivial, prices_per_order_trivial)
```

# Генетический алгоритм


```python
plt.figure(1)
plt.plot(orders_count_trivial, prices_trivial)
plt.scatter(orders_count_trivial, prices_trivial)
plt.plot(orders_count_genetic, prices_genetic)
plt.scatter(orders_count_genetic, prices_genetic)
plt.figure(2)
plt.plot(orders_count_trivial, prices_per_order_trivial)
plt.scatter(orders_count_trivial, prices_per_order_trivial)
plt.plot(orders_count_genetic, prices_per_order_genetic)
plt.scatter(orders_count_genetic, prices_per_order_genetic)
plt.figure(3)
plt.plot(orders_count_trivial, times_trivial)
plt.scatter(orders_count_trivial, times_trivial)
plt.plot(orders_count_genetic, times_genetic)
plt.scatter(orders_count_genetic, times_genetic)
```
